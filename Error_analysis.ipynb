```python
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
pd.options.mode.chained_assignment = None
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import scipy.stats as stats
```

```python
pred_df = pd.read_csv('predictions.csv')
```

```python
print(pred_df['y_pred'].describe())
print(pred_df['y_test'].describe())
```

```python
mse = mean_squared_error(pred_df['y_pred'], pred_df['y_test'])
mae = mean_absolute_error(pred_df['y_pred'], pred_df['y_test'])
r2 = r2_score(pred_df['y_pred'], pred_df['y_test'])

print('Mean Squared Error (MSE):', mse)
print('Mean Absolute Error (MAE):', mae)
print('R-squared:', r2)
```

```python
pred_df['residuals'] = pred_df['y_test'] - pred_df['y_pred']

plt.figure(figsize=(8, 6))
plt.hist(pred_df['residuals'], bins=20, color='darkcyan', edgecolor='black', alpha=0.7)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.grid(True)
plt.show()


sns.set(style="whitegrid")

sns.jointplot(x=pred_df['y_pred'], y=pred_df['residuals'], kind="kde", color="darkcyan", fill=True)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()
```

```python
stats.probplot(pred_df['residuals'], dist="norm", plot=plt)
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Quantiles from Prediction')
plt.title('QQ Plot of Residuals')
plt.grid(True)
plt.show()
```
